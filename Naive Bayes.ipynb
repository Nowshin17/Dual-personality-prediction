```python
import sklearn
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
import numpy as mp



from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB




```

```python
data = pd.read_csv(r"C:\Users\Ankita\Desktop\final\Naive Bayes\FinalData.csv")
```

```python
data.head(10)
```

```python
features = data.columns
#Display
features
```

```python
features = [x for x in features if x != 'Depressed']
features
```

```python
train, test = train_test_split(data, test_size = .25)
print(len(data))
print(len(train))
print(len(test))
```

```python
x_train = train[features]
y_train = train["Depressed"]

x_test = test[features]
y_test = test["Depressed"]
```

```python
Nb=GaussianNB()
Nb= Nb.fit(x_train,y_train)
```

```python
y_pred=Nb.predict(x_test)
```

```python
#Diplay
y_pred
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, y_pred) * 100
print("Accuracy using Naive Bayes: ", round(score, 1), "%" )
```
